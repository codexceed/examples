```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
v100_ai_x = np.linspace(0, 1000, 10000)
v100_flops = np.array(list(map(lambda a: min(a, 15.7 * 10**3), v100_ai_x * 900)))
v100_alexnet_ai = 16.76
v100_alexnet_gflops = 210.82
v100_resnet18_ai = 17.21
v100_resnet18_gflops = 739.53
v100_vgg16_ai = 15.7
v100_vgg16_gflops = 3978.17
```

```python
plt.plot(v100_ai_x, v100_flops, label="Roofline")
plt.plot([v100_alexnet_ai], [min(15.7 * 10**3, v100_alexnet_gflops)], "o", label="Alexnet")
plt.plot([v100_resnet18_ai], [min(15.7 * 10**3, v100_resnet18_gflops)], "o", label="Resnet18")
# plt.plot([v100_vgg16_ai], [min(15.7 * 10**3, v100_vgg16_gflops)], "o", label="VGG16")
plt.xlabel("Arithmetic Intensity (FLOPs/Byte)")
plt.ylabel("Bandwidth (GFLOPs/second)")
plt.xscale("log")
plt.yscale("log")
plt.legend(loc="best")
plt.title("NVIDIA V100 Roofline")
plt.savefig("v100_roofline.jpg")
```
